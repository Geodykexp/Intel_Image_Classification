# Adjusting the Learning rate

```python
def make_model(learning_rate=0.01):
    # 1. Load Base
    base_model = Xception(weights='imagenet',
                          include_top=False,
                          input_shape=(150, 150, 3))
    
    base_model.trainable = False

    # 2. Build Architecture
    inputs = keras.Input(shape=(150, 150, 3))
    # training=False ensures the batch norm layers stay in inference mode
    base = base_model(inputs, training=False)
    vectors = keras.layers.GlobalAveragePooling2D()(base)
    outputs = keras.layers.Dense(6, activation='softmax')(vectors)
    model = keras.Model(inputs, outputs)

    # 3. Compile
    optimizer = keras.optimizers.Adam(learning_rate=learning_rate)
    
    # from_logits=False because we have 'softmax' in the Dense layer
    loss = keras.losses.CategoricalCrossentropy(from_logits=False)

    model.compile(optimizer=optimizer,
                  loss=loss,
                  metrics=['accuracy'])

    return model
```

```python
# Dictionary to store the different learning rates
scores = {}

lrs = [0.0001, 0.001, 0.01,  0.1]

for lr in lrs:
    print(lr)

    model = make_model(learning_rate=lr)
    history = model.fit(train_ds, epochs=10, validation_data=test_ds)
    scores[lr] = history.history

print()
print()
```

```python
for lr, hist in scores.items():
    # plt.plot(hist['accuracy'], label=lr) #train
    plt.plot(hist['val_accuracy'], label=lr) #test
    plt.xticks(np.arange(10))
    plt.legend()
    
```

```python
#  It can be deduced that the best learning rate is 0.001
```

# Adding new layer - Relu

```python
# Function to define the model by adding a new dense layer
def make_model(learning_rate=0.01, size_inner=100): # default layer size is 100
    base_model = Xception(weights='imagenet',
                          include_top=False,
                          input_shape=(150,150,3))

    base_model.trainable = False

    #########################################

    inputs = keras.Input(shape=(150,150,3))
    base = base_model(inputs, training=False)
    vectors = keras.layers.GlobalAveragePooling2D()(base)
    
    # Adding another layer
    inner = keras.layers.Dense(size_inner, activation='relu')(vectors)

    # Adding drop_out
    # base = keras.layers.Dropout(0.5)(inner)
    # Randomly shuts off 50% of neurons during training

    # 4. Final Output (None, 6)
    outputs = keras.layers.Dense(6, activation='softmax')(inner)

    model = keras.Model(inputs, outputs)

    #########################################

    optimizer = keras.optimizers.Adam(learning_rate=learning_rate)
    loss = keras.losses.CategoricalCrossentropy(from_logits=True)

    # Compile the model
    model.compile(optimizer=optimizer,
                  loss=loss,
                  metrics=['accuracy'])

    return model
```

```python
# Experiment with different numbers of inner layers with the best learning rate
# Note: We should've added the checkpoint for training, but for simplicity, we are skipping it
learning_rate = 0.001

scores = {}

# List of inner layer sizes
sizes = [10, 100, 1000]

for size in sizes:
    print(size)

    model = make_model(learning_rate=learning_rate, size_inner=size)
    history = model.fit(train_ds, epochs=10, validation_data=test_ds)
    scores[size] = history.history

    print()
    print()
```

```python
for size, hist in scores.items():
    # plt.plot(hist['accuracy'], label=size) #'train'

    plt.plot(hist['val_accuracy'], label=('test_size=%s' % size)) #'test'
    #  or plt.plot(hist['val_accuracy'], label=f'test={size}')
    plt.xticks(np.arange(10))
    plt.yticks([0.75, 0.80, 0.85, 0.90, 0.95])
    plt.legend()
```

```python
# From the chart, it seems that with 10 epoch training, 
# Only size_inner = 100 tends to improve.
# While 10 and 1000 tend to reduce.
```

# Adding Dropout

```python
from tensorflow.keras.applications.xception import Xception
from tensorflow import keras

def make_model(learning_rate=0.01, size_inner=256, droprate=0.5):
    base_model = Xception(weights='imagenet',
                          include_top=False,
                          input_shape=(150, 150, 3))
    
    base_model.trainable = False

    ###############################################################

    inputs = keras.Input(shape=(150, 150, 3))
    base = base_model(inputs, training=False)
    vectors = keras.layers.GlobalAveragePooling2D()(base)
    
    # We use size_inner here
    inner = keras.layers.Dense(size_inner, activation='relu')(vectors)
    
    # We use droprate here
    drop = keras.layers.Dropout(droprate)(inner)
    
    outputs = keras.layers.Dense(6, activation='softmax')(drop)
    model = keras.Model(inputs, outputs)

    ###############################################################

    optimizer = keras.optimizers.Adam(learning_rate=learning_rate)
    loss = keras.losses.CategoricalCrossentropy(from_logits=False)

    model.compile(optimizer=optimizer,
                  loss=loss,
                  metrics=['accuracy'])

    return model
    

checkpoint = keras.callbacks.ModelCheckpoint(
    'xception_v2_{epoch:02d}_{val_accuracy:.3f}.keras', # Changed to .keras
    save_best_only=True,
    monitor='val_accuracy',
    mode='max'
)
```

```python
import tensorflow as tf
print("Num GPUs Available: ", len(tf.config.list_physical_devices('GPU')))
```

```python
# Set the best values of learning rate and inner layer size based on previous experiments
learning_rate = 0.001
size = 256

# Dict to store results
scores = {}

# List of dropout rates
droprates = [0.0, 0.2, 0.5, 0.8]

for droprate in droprates:
    print(droprate)

    model = make_model(learning_rate=learning_rate,
                       size_inner=size,
                       droprate=droprate)

    # Train for longer (epochs=30) cause of dropout regularization
    history = model.fit(train_ds, epochs=10, validation_data=test_ds, callbacks=[checkpoint])
    scores[droprate] = history.history

    print()
    print()
```

```python
for drop_rate, hist in scores.items():
    # plt.plot(hist['accuracy'], label=droprate) #'train'
    plt.plot(hist['val_accuracy'], label=('drop_rate=%s' % drop_rate)) #'test''
    plt.ylim()
    plt.xticks(np.arange(10))
    plt.legend()
```

```python
hist = scores[0.0]
plt.plot(hist['val_accuracy'], label = ('test=%s' % droprate ))
# plt.plot(hist['accuracy'], label = ('train=%s' % droprate ))

hist = scores[0.2]
plt.plot(hist['val_accuracy'], label = ('test=%s' % droprate ))
# plt.plot(hist['accuracy'], label = ('train=%s' % droprate ))

plt.legend()
```
